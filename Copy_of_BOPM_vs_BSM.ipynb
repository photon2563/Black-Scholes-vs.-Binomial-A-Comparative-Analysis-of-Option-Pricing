###Black Scholes Model
Black Scholes Model is a mathematical model which helps us to price options and other derivatives. This model is the first complete method to price options.
Before Black Scholes, option pricing was majorly a guesswork and traders used to rely on the intuition and the theory of demand and supply.
Black Scholes is based upon some assumptions as follow:
* **Log-Normal Distribution**: The stock prices follow Geometric Brownian motion with constant drift and volatility.
* **Constant Volatility**: The volatility of the asset remains constant througout the option's life.
* **Constant risk free rate**: The risk free interest rate remains constant.
* **No dividends**: The underlying stock pays no dividend during the lifetime of the option.
* **Perfect frictionless market**: The assets can be traded without any hindrance continuously.
* **European Exercise**: The options can only be exercised at the expiration.
* **No transaction costs**: No taxes, fees etc.
  
###**For call options we use**,
$C = S_0N(d_1) - Ke^{-rT}N(d_2)$
This can be understood as:
* $S_0N(d_1)$: This is the expected value which will be recieved when the option will be finish in-the-money.

* $Ke^{-rt}N(d_2)$: This part shows the present value of the strike price, if the option finishes in-the-money.

* $d_1$: It measures how far is the stock price from the strike price, adjusted for time and the volatility.

* $d_2$: Adjustment for risk free rate.

* $N(d_1)$: Normal dstribution term for the probability that the option will be exercised.

* $N(d_2)$: Risk neutral probability.

###**For put options we use**,
$P = Ke^{-rt}N(-d_2) - S_0N(-d_1)$

Where:
$d_1=\frac{[ln(S_0/K) + (r + σ^{2}/2)T] }{ (σ\sqrt(T))}$

$d_2=d_1 - σ\sqrt(T)$

Parameters:
$C$ = Call option price

$P$ = Put option price

$S₀$ = Current stock price

$K$ = Strike price

$r$ = Risk free interest rate

$T$ = Time to expiration

$σ$ = Volatility of the asset

$N(x)$ = Cumulative standard normal distribution function

$e$ = Euler's number

$ln$ = Natural logarithm

###Binomial Option Pricing Model

Binomial Option Pricing Model (BOPM) is a method used to estimate the price of an option. Unlike the Black Scholes Model, the BOPM basically breaks down the life of an option into a series of upwards and downward motions.
* Up(u) = The price increases by this factor.
* Down(d) = The price decreases by this factor.

**Risk neutral probabilities**:
* $ p = \frac{(e^{rΔt} - d)}{(u-d)}$

* $ 1-p = \frac{(u - e^{rΔt})}{(u-d)}$

Where r is the risk free rate and Δt is the time step length.

Parameters:
* $ u = e^{σ\sqrt(Δt)}$
* $ d = e^{-σ\sqrt(Δt)}$    or   $ d = \frac{1}{u}$
* $σ$ = volatility
* $r$ = risk free rate
* $T$ = time to expiration
* $n$ = number of time steps

#Black Scholes Model
---

###**The Constant Risk-Free Rate Assumption**
Risk-free rate is the return the return the investor would earn by putting their money somewhere it is completely safe, like lending it to the government. The Black Scholes Model assumes the existence of only a single constant risk-free rate for the entire life of the option, which is assumed to be a hypothetical return that the investor has the certainty of getting.  In reality, risk-free rates are not constant over the course; they keep on changing as a consequence of multiple things, including policy changes, any market sentiment shift, forecasts and a lot more real-world happenings. For example, the famous 2008 financial crisis or anything like the COVID-19 pandemic. So, keeping these factors in mind, certainly, constant risk-free rates can result in extreme mispricing and inappropriate option prices for the stock under consideration. This assumption is not very suitable for the longer-term options. Also, the model offers infinite lending and borrowing at the risk-free rate, which is not possible for the major chunk of the market participants.

###**Geometric Brownian Motion**

This assumption allows modelling of the process to be random, yet continuous, using differential equations. It makes the process tractable.
Implications:

•	Returns turn out to be normally distributed.

•	Asset prices are log-normally distributed, so they are always positive.

•	No sudden jumps.

The model assumption about GBM is a very fascinating feature, but also a practical flaw. In this, we assume that the price movements are random and independent. Also, it is assumed that it is normally distributed (in log terms). So what it means is that today’s price movements provide no information about any movement in future, which basically means it is memoryless. But practically, in the longer period, it fails to fit right there. Real stock prices exhibit momentum effects, where trends tend to continue over short terms and eventually return to the fundamental values over the longer periods. GBM also suggest that extreme price movements have very little probability of happening. The 1987 stock market crash is an example of that, and obviously, extreme price movements are more often than what this model suggests.

###**Constant Volatility**
Volatility is the size of price variations, and the model suggests that it doesn’t vary, no matter the underlying market conditions, remaining time to expiry, or strike price compared to the current market price.

Implication:

•	The model ignores any changes in volatility due to real-world interventions, such as market sentiments and financial reports.

This preserves the fact that a sudden change in volatility can’t take place. Contrary to that, in 2008, the volatility index spiked to levels above 80% compared to the usual 15-20%. If the Black-Scholes model held true, the options that are on the same underlying asset with the same maturity date should have the same volatility, which is not the case.

###**The No Dividend assumption**
Implication:

  •	This model limits the underlying asset from paying any dividends to the investors.

This assumption suggests that there is no dividend payout throughout the life of the option. This is not really true because many options are traded actively to make regular dividend payments. Dividends have a direct influence on the option values. A stock value falls by the amount of the dividend when it becomes ex-dividend. In a call option, it reduces the value, since it reduces the benefit that can be utilised owning the stock, and for this same reason, the value increases in the case of a put option.

These models try to accommodate the dividends by considering them as payments/yields, but doing so makes the model less easy to use and requires a good dividend forecast.


###**European Exercise Restriction**
This model only provides a formulation for European-style options. This is so because the model doesn’t sit well with the options that provide early exercise opportunities. This is the reason why Black Scholes model leads to an underestimation of American options by a large number.

Early exercise is useful for the dividend-paying stocks. For put options, early exercise is a very optimised way. When the stock value depreciates by a significant amount, the holder can opt for early exercise and deposit the proceeds at the risk-free rate, rather than waiting for the option to expire. Such a thing is not allowed in Black Scholes model, making it not so efficient for American options.

###**No Arbitrage**
This assumption emphasises the fact that there is no riskless profit. One can’t opt for two different markets at the same time.

The Black Scholes model assumes that there aren’t even momentary arbitrage opportunities. This is so because BSM assumes that the market is like a fluid system, and therefore any small discrepancy in the pricing would lead to trading activities that will restore equilibrium. This basically means that under the domain of BSM, it is impossible for any arbitrage opportunity to exist.

Well, this is all because the Black Scholes model is a continuous time model.

###**Frictionless Market**
This BSM model assumes that continuous trading is always possible without any restrictions or market constraints. Basically, infinite liquidity.

The truth is something else. In the actual market, there are transaction costs associated, along with exchange fees, and some other regulatory charges. If that had been a case for the BSM model, which provides a continuous time framework, then it would have been an infinite transaction fee, since it allows any number of transaction in a timeframe.


#Binomial Options Pricing Model
---
###**Discrete Time Price Movement**
The BOPM is laid on the foundation of the movement of the price up and down by constant factors at each of the steps. This assumption is helpful because it makes the model more tractable and allows a clear visualisation of the price movement in the time frame. It is assumed that at each step, the price moves up by a factor of u and down by d.

But in the actual market, the price movement is much more dynamic, governed by multiple factors like market sentiments, order imbalances, etc. And therefore, the restriction applied to the price movement, allowing only two possibilities, removes all those probable realities. Also, since there are only two possible price movements in BOPM, it provides only some specific pathways to hit a price, whereas in the real market, there are multiple ways to reach the same price point, via different volatilities and timings.

###**Constant Up and Down factors**
The binomial model assumes that the price movements have constant multiplicative factors throughout the complete course of the option’s life. That means if the stock can move up by a factor of 5% or down by 10%, these percentages do not vary for market shifts and other real time conditions.
This assumption certainly helps by making the process simpler, but it is obviously inconsistent with the facts. Anything like the annual reports or merger announcements can significantly create a change in the market. This constant assumption is associated with the particular volatility structure that is not compatible with the actual market conditions.
This assumption also introduces symmetry in the volatility that may not be present in the actual market. Most of the stocks have asymmetric volatility, particularly the downward movements are more volatile than the upward movements. This proves the fact that bad news travels at a faster rate than good news.

###**Risk neutral probability assumption**
The model is based upon risk neutral probabilities that show no change for any market situation. They are derived so that the expected return is equal to the risk free rate.
Although this assumption yields mathematical convenience and arbitrage-free prices, it is actually a disconnect from the actual market. In reality, the probabilities vary due to multiple factors, including market mood, economic tensions, financial data and investors’ risk tolerance, etc. In a bearish market, the probability of a downward move increases, whereas in a bullish market, we see the opposite.

###***Also these model’s probabilities serve only pricing purposes but they do not actually provide any meaningful information about the expected outcomes.***

###***Alongside all this, BOPM is flexible enough to accommodate early exercise decisions at each node, making it somewhat suitable for American options.***

## Market Perfection Assumptions
---
###**Frictionless Trading**

The two models have frictionless markets where trades are executed at zero latencies, costs, or market impact. The assumption eliminates bid-ask spreads, exchange commissions, brokerage commissions, and market impact costs that characterize real trading conditions.

In reality, option strategies are significantly impacted by transaction costs, especially for costless frequent rebalancing strategies hypothesized under the models. The Black Scholes model delta hedging approach is to rebalance frequently in order to have perfect hedges, but with each rebalancing transaction comes an expense that adds up over time. For the individual investor, these transaction costs are enough to render theoretically profitable strategies economically unfeasible.

bid-ask spreads are also harder to manage, particularly for illiquid options whose spreads range several percentage points wide. The theoretical option value calculated by either model will lie between the bid and offer prices and cannot be bought or sold at the theoretical fair value. The spread offers a range of acceptable prices rather than the point estimate that the models yield.

Market impact comes into play for large option positions, where attempting to open or close positions will move market prices against you. Infinite liquidity assumption in models did not take into account the fact that large trades can move prices away from theoretical fair value, with costs not captured in the mathematical models.
###**Perfect Borrowing and Lending**

Both models entail investors lending and borrowing unlimited amounts at the risk-free rate to create perfect hedge strategies and arbitrage positions. It is on this basis that the models are capable of creating risk-free portfolios with dynamic hedging.

Reality places severe lending and borrowing constraints that shatter such hopes. Private investors are typically unable to borrow at rates on or close to government bond rates, with very large spreads covering credit risk and funding capital requirements. Institutional investors are more fortunate, but they too face restraint in times of market stress when liquidity evaporates and borrowing is costly.

Short-selling limitations render the perfect borrowing assumption even more unrealistic. Borrowing is expensive or infeasible for shortable stocks in general and stocks can remain unshortable when demand is high. These limitations guarantee that perfect arbitrage relations between the two models are non-existent, which provides scope for ongoing mispricing that cannot be explained by the models.

Margin requirements add another level of sophistication because investors must put up capital-efficiency-hungry collateral with opportunity costs. Costlessness assumptions of carrying positions in models ignore such capital efficiency problems that afflict real implementation.

###**Continuous Trading**

These two models also require that continuous trading is possible in order to facilitate continuous rebalancing required by optimal hedge strategies. This ignores the fact that most financial markets trade at specific times with routine weekly and holiday closings.

Market closures expose price-hedging gaps to which investors are subjected to substantial weekend and overnight risk. Major news announcements, earnings releases, and international news typically occur when markets are closed, creating price gaps in contrast with the assumption of continuous trading. Such gaps have the potential to push established hedges out of equilibrium and induce losses that the models are not anticipating.

The assumption is most questionable in times of stress in the markets when trades can stop in a particular stock or overall market. Circuit breakers, trading halts, and excessive volatility can prevent ongoing rebalancing from happening, something both models implicitly assume.

In addition, different trading sessions and time zones allow arbitrage opportunities and price mismatches that are not possible via trading. Foreign index or stock options raise particular problems if underlying markets are of other hours than options markets.

## Advanced Implications and Model Interactions
----

###**Volatility Surface Dynamics**

The two models' volatility hypothesis of continually fluctuating volatility creates a built-in contradiction with reality in the behavior of the implied volatilities to move systematically with strike price and expiration date. Such movement creates the volatility surface.

The volatility smile will typically display higher implied volatilities for the out-of-the-money options, or the price that investors are willing to pay a premium to have coverage in the tail. The profile is the reverse of the model's assumption that all the options reflect the same volatility. The skew will normally be downward-sloping with out-of-the-money puts demanding higher implied volatilities than calls, suggesting that the market is more likely to see large moves down than the normal distribution.

Time-varying patterns of volatility add another layer of complexity. Implied volatilities increase as earnings announcements or other anticipated events approach and subsequently decline once uncertainty has been alleviated. Such a volatility term structure presents opportunities and risks that constant volatility models cannot price.

The dynamics in the volatility surface pose hedging and risk management concerns. Positions that are optimally hedged with the assumption of stationary volatility can turn out to be colossal volatility risk when the surface is stochastically in motion. Seasoned practitioners will put the volatility surface models on top of the underlying models to have a better representation of the dynamics.

###**Model Risk and Parameter Sensitivity**

Both models are highly sensitive to input parameters, particularly to volatility, and therefore bear enormous model risk when the parameters are computed in error. Small changes in assumed volatility yield enormous differences in estimated option prices, for instance, on long term or deep out-of-the-money options.

This sensitivity also creates issues for practitioners, who have to estimate future volatility based on history, implied volatilities, or judgment. History will not always be the future, and implied volatilities reflect market sentiment and risk premiums that might be non-fair value estimates.

The model sensitivity to the assumptions of interest rates is especially important when there is volatility in the rate. Although interest rates will have less effect on short options, longer options can be extremely sensitive to rates of change, posing additional parameter estimation problems.

Correlation assumptions, though not required in single asset models but required when the models are applied for multi asset derivatives or portfolios, go against the independent price movement assumption in the models if the assets are correlated and have systematic risks which cannot be identified by single asset models.

###**Convergence Properties and Numerical Accuracy**

The approach of the binomial model to the Black-Scholes solution is reassuring on a theoretical basis but problematic in reality. Convergence will be slow and oscillating, and one has to choose the number of time steps cautiously so that reasonable accuracy is attained.

The decision of the up and down parameters determines quality and rate of convergence. The Cox-Ross-Rubinstein parameterization is well behaved and convergent but need not be the best currently available for every application. There exist parameterizations which can provide higher quality for some applications at the cost of interpretability of the model.

Numerical precision is all the more vital in the case of multifactor derivatives or option sensitivities (Greeks). Small variations in option prices can cause large errors in hedge ratios calculated, and thereby astronomical trading losses on entry of erroneous hedging strategies.

Computation that goes into achieving high accuracy may be time-consuming, particularly in the American options where computation of the early exercise cost needs to be done at each node. The computation limits application of the model in real-time or exotic derivatives where computation has to be fast.

## 1. Framework

### Binomial Model Parameters

Let us establish the binomial model with the following parameters:

$$\Delta t = \frac{T}{n} \quad \text{(time step length)} \tag{1}$$

$$u = e^{\sigma\sqrt{\Delta t}} \quad \text{(up factor)} \tag{2}$$

$$d = e^{-\sigma\sqrt{\Delta t}} = \frac{1}{u} \quad \text{(down factor)} \tag{3}$$

$$p = \frac{e^{r\Delta t} - d}{u - d} \quad \text{(risk-neutral probability)} \tag{4}$$

### Black-Scholes Framework

The continuous-time model assumes the asset price follows (assumption of GBM):

$$dS_t = rS_t dt + \sigma S_t dW_t \tag{5}$$

with solution:

$$S_T = S_0 \exp\left(\left(r - \frac{\sigma^2}{2}\right)T + \sigma W_T\right) \tag{6}$$

---

## 2. Step 1: Asymptotic Analysis of Parameters

### **Lemma 1** (Parameter Asymptotics)

As $n \to \infty$ (equivalently $\Delta t \to 0$), we have:

$$u = 1 + \sigma\sqrt{\Delta t} + \frac{\sigma^2\Delta t}{2} + O((\Delta t)^{3/2}) \tag{7}$$

$$d = 1 - \sigma\sqrt{\Delta t} + \frac{\sigma^2\Delta t}{2} + O((\Delta t)^{3/2}) \tag{8}$$

$$p = \frac{1}{2} + \frac{r\sqrt{\Delta t}}{2\sigma} + O(\Delta t) \tag{9}$$

###  **Proof:**

Using Taylor expansion of $e^x = 1 + x + \frac{x^2}{2} + O(x^3)$:

**For the up factor:**
$$u = e^{\sigma\sqrt{\Delta t}} = 1 + \sigma\sqrt{\Delta t} + \frac{(\sigma\sqrt{\Delta t})^2}{2} + O((\Delta t)^{3/2})$$
$$= 1 + \sigma\sqrt{\Delta t} + \frac{\sigma^2\Delta t}{2} + O((\Delta t)^{3/2})$$

**For the probability:**
$$p = \frac{e^{r\Delta t} - d}{u - d} = \frac{(1 + r\Delta t) - (1 - \sigma\sqrt{\Delta t})}{2\sigma\sqrt{\Delta t}} + \text{higher order terms}$$
$$= \frac{1}{2} + \frac{r\sqrt{\Delta t}}{2\sigma} + O(\Delta t)$$

---

## 3. Step 2: Log-Return Analysis

### **Definition 1** (Binomial Log-Returns)

Let $X_i$ be the $i$-th step log-return in the binomial model:

$$X_i = \begin{cases}
\ln(u) = \sigma\sqrt{\Delta t} & \text{with probability } p\\
\ln(d) = -\sigma\sqrt{\Delta t} & \text{with probability } 1-p
\end{cases} \tag{10}$$

The total log-return over $n$ steps is:
$$\log\left(\frac{S_T}{S_0}\right) = \sum_{i=1}^n X_i \tag{11}$$

### **Lemma 2** (Moment Analysis)

The first two moments of $X_i$ are:

$$E[X_i] = (2p-1)\sigma\sqrt{\Delta t} = r\Delta t \tag{12}$$

$$\text{Var}[X_i] = \sigma^2\Delta t - (E[X_i])^2 = \sigma^2\Delta t + O((\Delta t)^2) \tag{13}$$

### **Proof:**

From equation (9):
$$E[X_i] = (2p-1)\sigma\sqrt{\Delta t} = \left(2\left(\frac{1}{2} + \frac{r\sqrt{\Delta t}}{2\sigma}\right) - 1\right)\sigma\sqrt{\Delta t} = r\Delta t$$

For the variance:
$$E[X_i^2] = p(\sigma\sqrt{\Delta t})^2 + (1-p)(\sigma\sqrt{\Delta t})^2 = \sigma^2\Delta t$$
$$\text{Var}[X_i] = \sigma^2\Delta t - (r\Delta t)^2 = \sigma^2\Delta t + O((\Delta t)^2)$$

---

## 4. Step 3: Central Limit Theorem Application

### **Theorem 1** (Convergence in Distribution)

As $n \to \infty$, the standardized sum of log-returns converges to a standard normal distribution:

$$\frac{\sum_{i=1}^n X_i - n \cdot E[X_i]}{\sqrt{n \cdot \text{Var}[X_i]}} \xrightarrow{d} N(0,1) \tag{14}$$

### **Proof:**

From Lemma 2:
- $n \cdot E[X_i] = n \cdot r\Delta t = nr\frac{T}{n} = rT$
- $n \cdot \text{Var}[X_i] = n \cdot \sigma^2\Delta t = n\sigma^2\frac{T}{n} = \sigma^2T$

Therefore:
$$\frac{\sum_{i=1}^n X_i - rT}{\sigma\sqrt{T}} \xrightarrow{d} N(0,1)$$

This implies:
$$\sum_{i=1}^n X_i \xrightarrow{d} N(rT, \sigma^2T) \tag{15}$$

---

## 5. Step 4: Risk-Neutral Measure Adjustment

### **Proposition 1** (Risk-Neutral Convergence)

Under the risk-neutral measure, the log-return converges to:

$$\log\left(\frac{S_T}{S_0}\right) \xrightarrow{d} N\left(\left(r - \frac{\sigma^2}{2}\right)T, \sigma^2T\right) \tag{16}$$

### **Proof:**

We need to account for the **Itô correction term**. Using Itô's lemma on $\log(S_t)$ where $dS_t = rS_t dt + \sigma S_t dW_t$:

$$d(\log S_t) = \frac{1}{S_t}dS_t - \frac{1}{2S_t^2}(dS_t)^2$$
$$= rdt + \sigma dW_t - \frac{\sigma^2}{2}dt = \left(r - \frac{\sigma^2}{2}\right)dt + \sigma dW_t$$

Integrating from $0$ to $T$:
$$\log(S_T) - \log(S_0) = \left(r - \frac{\sigma^2}{2}\right)T + \sigma W_T$$

Since $W_T \sim N(0,T)$, we obtain the desired result.

---

## 6. Step 5: Option Pricing Convergence

###  **Theorem 2** (Main Convergence Result)

For a European call option with strike $K$, the binomial option price converges to the Black-Scholes price:

$$\boxed{\lim_{n \to \infty} C_{\text{BOPM}}^{(n)} = C_{\text{BS}}} \tag{17}$$

where:
$$C_{\text{BS}} = S_0\Phi(d_1) - Ke^{-rT}\Phi(d_2) \tag{18}$$

$$d_1 = \frac{\ln(S_0/K) + (r + \sigma^2/2)T}{\sigma\sqrt{T}} \tag{19}$$

$$d_2 = d_1 - \sigma\sqrt{T} \tag{20}$$

### **Proof:**

The binomial option price is:
$$C_{\text{BOPM}}^{(n)} = e^{-rT}E_{\mathbb{Q}}\left[\max(S_T - K, 0)\right]$$

where $S_T = S_0\exp\left(\sum_{i=1}^n X_i\right)$ and the expectation is under the risk-neutral measure.

From Proposition 1, as $n \to \infty$:
$$S_T \xrightarrow{d} S_0\exp\left(N\left(\left(r - \frac{\sigma^2}{2}\right)T, \sigma^2T\right)\right)$$

This is exactly the terminal stock price distribution in the Black-Scholes model.

By the **continuous mapping theorem** and **dominated convergence theorem**, we have:
$$\lim_{n \to \infty} C_{\text{BOPM}}^{(n)} = e^{-rT}E_{\mathbb{Q}}\left[\max(S_T^{\text{BS}} - K, 0)\right] = C_{\text{BS}}$$

---

## 7. Step 6: Rate of Convergence

### **Proposition 2** (Convergence Rate)

The convergence rate is of order $O(1/\sqrt{n})$:

$$\left|C_{\text{BOPM}}^{(n)} - C_{\text{BS}}\right| = O\left(\frac{1}{\sqrt{n}}\right) \tag{21}$$

This follows from the **Berry-Esseen theorem**, which provides bounds on the rate of convergence in the Central Limit Theorem.

```python
import numpy as np
import matplotlib.pyplot as plt
import yfinance as yf
from scipy.stats import norm
from scipy.optimize import brentq
from IPython.display import display, clear_output
import ipywidgets as widgets

# Black-Scholes Model
class BlackScholes:
    def __init__(self, spot_price, strike, maturity, risk_free_rate, volatility, option_type='call'):
        self.S = spot_price
        self.K = strike
        self.T = maturity
        self.r = risk_free_rate
        self.sigma = volatility
        self.option_type = option_type.lower()

    def d1(self):
        return (np.log(self.S / self.K) + (self.r + 0.5 * self.sigma ** 2) * self.T) / (self.sigma * np.sqrt(self.T))

    def d2(self):
        return self.d1() - self.sigma * np.sqrt(self.T)

    def price(self):
        d1 = self.d1()
        d2 = self.d2()
        if self.option_type == 'call':
            return self.S * norm.cdf(d1) - self.K * np.exp(-self.r * self.T) * norm.cdf(d2)
        else:
            return self.K * np.exp(-self.r * self.T) * norm.cdf(-d2) - self.S * norm.cdf(-d1)

    def delta(self):
        d1 = self.d1()
        return norm.cdf(d1) if self.option_type == 'call' else norm.cdf(d1) - 1

    def gamma(self):
        d1 = self.d1()
        return norm.pdf(d1) / (self.S * self.sigma * np.sqrt(self.T))

    def theta(self):
        d1 = self.d1()
        d2 = self.d2()
        term1 = - (self.S * norm.pdf(d1) * self.sigma) / (2 * np.sqrt(self.T))
        if self.option_type == 'call':
            term2 = self.r * self.K * np.exp(-self.r * self.T) * norm.cdf(d2)
            return term1 - term2
        else:
            term2 = self.r * self.K * np.exp(-self.r * self.T) * norm.cdf(-d2)
            return term1 + term2

    def vega(self):
        d1 = self.d1()
        return self.S * norm.pdf(d1) * np.sqrt(self.T)

    def rho(self):
        d2 = self.d2()
        if self.option_type == 'call':
            return self.K * self.T * np.exp(-self.r * self.T) * norm.cdf(d2)
        else:
            return -self.K * self.T * np.exp(-self.r * self.T) * norm.cdf(-d2)

# Binomial Option Pricing Model
def binomial_tree_price(S, K, T, r, sigma, steps, option_type='call'):
    dt = T / steps
    u = np.exp(sigma * np.sqrt(dt))
    d = 1 / u
    q = (np.exp(r * dt) - d) / (u - d)

    ST = np.array([S * (u ** j) * (d ** (steps - j)) for j in range(steps + 1)])
    option_values = np.maximum(ST - K, 0) if option_type == 'call' else np.maximum(K - ST, 0)

    for i in range(steps, 0, -1):
        option_values = np.exp(-r * dt) * (q * option_values[1:] + (1 - q) * option_values[:-1])

    return option_values[0]

# Implied Volatility Calculator (Brent's method)
def implied_volatility(target_price, S, K, T, r, option_type='call'):
    def objective(sigma):
        if sigma <= 0:
            return target_price
        bs = BlackScholes(S, K, T, r, sigma, option_type)
        return bs.price() - target_price
    try:
        return brentq(objective, 1e-5, 5)
    except ValueError:
        return np.nan

# Live price fetcher
def get_live_price(ticker):
    stock = yf.Ticker(ticker)
    return stock.history(period="1d")['Close'].iloc[-1]

# Plotting convergence of BOPM to BSM
def plot_models(S, K, T, r, sigma, steps, option_type='call'):
    bs_price = BlackScholes(S, K, T, r, sigma, option_type).price()
    binomial_prices = [binomial_tree_price(S, K, T, r, sigma, n, option_type) for n in range(1, steps + 1)]
    plt.figure(figsize=(10, 6))
    plt.plot(range(1, steps + 1), binomial_prices, label='Binomial Tree Price')
    plt.axhline(bs_price, color='red', linestyle='--', label='Black-Scholes Price')
    plt.title(f'BOPM vs Black-Scholes for {option_type.capitalize()} Option')
    plt.xlabel('Number of Steps in Binomial Tree')
    plt.ylabel('Option Price')
    plt.legend()
    plt.grid(True)
    plt.show()

# Main interaction function
def interactive_pricing(ticker, K, T, r, sigma, steps, option_type):
    with out:
        clear_output(wait=True)
        S = get_live_price(ticker)
        bs_model = BlackScholes(S, K, T, r, sigma, option_type)
        bs_price = bs_model.price()
        bopm_price = binomial_tree_price(S, K, T, r, sigma, steps, option_type)
        implied_vol = implied_volatility(bopm_price, S, K, T, r, option_type)

        print(f"\nLive Stock Price of {ticker.upper()}: ${S:.2f}")
        print(f"\nOption Type: {option_type.capitalize()}")
        print(f"Strike Price: {K}")
        print(f"Time to Maturity: {T} years")
        print(f"Risk-Free Rate: {r * 100:.2f}%")
        print(f"Volatility (Input): {sigma:.2f}")
        print(f"Steps in BOPM: {steps}")
        print("\n--- Prices ---")
        print(f"Black-Scholes Price: ${bs_price:.4f}")
        print(f"Binomial Tree Price: ${bopm_price:.4f}")
        print(f"Implied Volatility (from BOPM Price): {implied_vol:.4f}")
        print("\n--- Greeks (from BSM) ---")
        print(f"Delta: {bs_model.delta():.4f}")
        print(f"Gamma: {bs_model.gamma():.4f}")
        print(f"Theta: {bs_model.theta():.4f}")
        print(f"Vega : {bs_model.vega():.4f}")
        print(f"Rho  : {bs_model.rho():.4f}")

        plot_models(S, K, T, r, sigma, steps, option_type)

# Widgets
ticker_widget = widgets.Text(value='NVDA', description='Ticker:')
K_widget = widgets.FloatSlider(value=100, min=50, max=300, step=1, description='Strike:')
T_widget = widgets.FloatSlider(value=1, min=0.01, max=2, step=0.01, description='Time (yrs):')
r_widget = widgets.FloatSlider(value=0.05, min=0.0, max=0.1, step=0.001, description='Risk-Free Rate:')
sigma_widget = widgets.FloatSlider(value=0.5, min=0.01, max=1.0, step=0.01, description='Volatility:')
steps_widget = widgets.IntSlider(value=100, min=1, max=1000, step=1, description='Steps:')
option_type_widget = widgets.Dropdown(options=['call', 'put'], value='call', description='Option Type:')

ui = widgets.VBox([
    ticker_widget, K_widget, T_widget, r_widget, sigma_widget, steps_widget, option_type_widget
])
out = widgets.Output()

display(ui, out)

widgets.interactive_output(
    interactive_pricing,
    {
        'ticker': ticker_widget,
        'K': K_widget,
        'T': T_widget,
        'r': r_widget,
        'sigma': sigma_widget,
        'steps': steps_widget,
        'option_type': option_type_widget
    }
)
```

##**Below are the plots of different parameters against option price, including strike price, volatility, risk-free rate and time to maturity. Also, there are interactive widgets to switch between call and put options, one for switching between the parameters *(strike price, volatility, risk-free rate and time to maturity)* and other for switching between the models, i.e. BOPM and BSM. There is also plots showing the sensitivity of greeks against these very parameters. For BOPM, the greeks calculation is done using finite difference.**



```python
import warnings
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import Dropdown, interact
from scipy.stats import norm

warnings.filterwarnings("ignore")


# --- Black Scholes Model ---
def d1(S, K, T, r, sigma):
    return (np.log(S / K) + (r + 0.5 * sigma**2) * T) / (sigma * np.sqrt(T))


def d2(S, K, T, r, sigma):
    return d1(S, K, T, r, sigma) - sigma * np.sqrt(T)


def black_scholes(S, K, T, r, sigma, option_type="call"):
    D1, D2 = d1(S, K, T, r, sigma), d2(S, K, T, r, sigma)
    if option_type == "call":
        return S * norm.cdf(D1) - K * np.exp(-r * T) * norm.cdf(D2)
    return K * np.exp(-r * T) * norm.cdf(-D2) - S * norm.cdf(-D1)


def delta_bs(S, K, T, r, sigma, option_type="call"):
    D1 = d1(S, K, T, r, sigma)
    return norm.cdf(D1) if option_type == "call" else norm.cdf(D1) - 1


def gamma_bs(S, K, T, r, sigma):
    D1 = d1(S, K, T, r, sigma)
    return norm.pdf(D1) / (S * sigma * np.sqrt(T))


def vega_bs(S, K, T, r, sigma):
    D1 = d1(S, K, T, r, sigma)
    return S * norm.pdf(D1) * np.sqrt(T)


def theta_bs(S, K, T, r, sigma, option_type="call"):
    D1, D2 = d1(S, K, T, r, sigma), d2(S, K, T, r, sigma)
    term1 = -(S * norm.pdf(D1) * sigma) / (2 * np.sqrt(T))
    if option_type == "call":
        return term1 - r * K * np.exp(-r * T) * norm.cdf(D2)
    return term1 + r * K * np.exp(-r * T) * norm.cdf(-D2)


def rho_bs(S, K, T, r, sigma, option_type="call"):
    D2 = d2(S, K, T, r, sigma)
    if option_type == "call":
        return K * T * np.exp(-r * T) * norm.cdf(D2)
    return -K * T * np.exp(-r * T) * norm.cdf(-D2)


# --- Binomial Option Pricing Model ---
def binomial_option_price(S, K, T, r, sigma, option_type="call", N=100):
    dt = T / N
    u = np.exp(sigma * np.sqrt(dt))
    d = 1 / u
    p = (np.exp(r * dt) - d) / (u - d)

    # Initialize asset prices and option values at maturity
    ST = np.array([S * (u**j) * (d ** (N - j)) for j in range(N + 1)])
    option_values = (
        np.maximum(ST - K, 0) if option_type == "call" else np.maximum(K - ST, 0)
    )

    # Backward induction
    for i in range(N - 1, -1, -1):
        option_values = np.exp(-r * dt) * (
            p * option_values[1:] + (1 - p) * option_values[:-1]
        )
    return option_values[0]


# --- Greeks via finite differences for BOPM ---
def delta_bopm(S, K, T, r, sigma, option_type="call", N=100, h=0.5):
    price_up = binomial_option_price(S + h, K, T, r, sigma, option_type, N)
    price_down = binomial_option_price(S - h, K, T, r, sigma, option_type, N)
    return (price_up - price_down) / (2 * h)

def gamma_bopm(S, K, T, r, sigma, option_type="call", N=100, h=0.5):
    price_up = binomial_option_price(S + h, K, T, r, sigma, option_type, N)
    price = binomial_option_price(S, K, T, r, sigma, option_type, N)
    price_down = binomial_option_price(S - h, K, T, r, sigma, option_type, N)
    return (price_up - 2 * price + price_down) / (h**2)

def vega_bopm(S, K, T, r, sigma, option_type="call", N=100, h=0.01):
    price_up = binomial_option_price(S, K, T, r, sigma + h, option_type, N)
    price_down = binomial_option_price(S, K, T, r, sigma - h, option_type, N)
    return (price_up - price_down) / (2 * h)

def theta_bopm(S, K, T, r, sigma, option_type="call", N=100, h=1 / 365):
    if T - h <= 0:
        return np.nan
    price_up = binomial_option_price(S, K, T, r, sigma, option_type, N)
    price_down = binomial_option_price(S, K, T - h, r, sigma, option_type, N)
    return (price_down - price_up) / h

def rho_bopm(S, K, T, r, sigma, option_type="call", N=100, h=0.0001):
    price_up = binomial_option_price(S, K, T, r + h, sigma, option_type, N)
    price_down = binomial_option_price(S, K, T, r - h, sigma, option_type, N)
    return (price_up - price_down) / (2 * h)

# --- Core Logic Bridge ---
def price_and_greeks(
    S, K, T, r, sigma, option_type, model="Black-Scholes", N=100
):
    if model == "Black-Scholes":
        return (
            black_scholes(S, K, T, r, sigma, option_type),
            delta_bs(S, K, T, r, sigma, option_type),
            gamma_bs(S, K, T, r, sigma),
            vega_bs(S, K, T, r, sigma),
            theta_bs(S, K, T, r, sigma, option_type),
            rho_bs(S, K, T, r, sigma, option_type),
        )
    return (
        binomial_option_price(S, K, T, r, sigma, option_type, N),
        delta_bopm(S, K, T, r, sigma, option_type, N),
        gamma_bopm(S, K, T, r, sigma, option_type, N),
        vega_bopm(S, K, T, r, sigma, option_type, N),
        theta_bopm(S, K, T, r, sigma, option_type, N),
        rho_bopm(S, K, T, r, sigma, option_type, N),
    )


# --- Interactive Plot Execution ---
def plot_price_and_greeks(
    option_type="call", parameter="Strike Price", model="Black-Scholes"
):
    base = {"S": 100, "K": 100, "T": 1.0, "r": 0.05, "sigma": 0.2}
    param_ranges = {
        "Strike Price": np.linspace(50, 150, 50),
        "Volatility": np.linspace(0.01, 1.0, 50),
        "Time to Maturity": np.linspace(0.01, 2.0, 50),
        "Risk-Free Rate": np.linspace(0.0, 0.2, 50),
    }
    x = param_ranges[parameter]

    price_list, delta_list, gamma_list, vega_list, theta_list, rho_list = (
        [],
        [],
        [],
        [],
        [],
        [],
    )

    for val in x:
        S, K, T, r, sigma = (
            base["S"],
            base["K"],
            base["T"],
            base["r"],
            base["sigma"],
        )
        if parameter == "Strike Price":
            K = val
        elif parameter == "Volatility":
            sigma = val
        elif parameter == "Time to Maturity":
            T = val
        elif parameter == "Risk-Free Rate":
            r = val

        p, d, g, v, th, rh = price_and_greeks(
            S, K, T, r, sigma, option_type, model
        )
        price_list.append(p)
        delta_list.append(d)
        gamma_list.append(g)
        vega_list.append(v)
        theta_list.append(th)
        rho_list.append(rh)

    # Visualization Setup
    fig, axs = plt.subplots(3, 2, figsize=(14, 12))
    fig.suptitle(
        f"{model} {option_type.capitalize()} Option Sensitivity to {parameter}",
        fontsize=16,
    )

    plots = [
        (axs[0, 0], price_list, "Option Price", "C0"),
        (axs[0, 1], delta_list, "Delta", "orange"),
        (axs[1, 0], gamma_list, "Gamma", "green"),
        (axs[1, 1], vega_list, "Vega", "red"),
        (axs[2, 0], theta_list, "Theta", "purple"),
        (axs[2, 1], rho_list, "Rho", "brown"),
    ]

    for ax, data, label, color in plots:
        ax.plot(x, data, label=label, color=color)
        ax.set_title(label)
        ax.grid(True)
        ax.set_xlabel(parameter)
        ax.set_ylabel("Value")
        ax.legend()

    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    plt.show()


# --- Display Widgets ---
interact(
    plot_price_and_greeks,
    option_type=Dropdown(
        options=["call", "put"], value="call", description="Option Type:"
    ),
    parameter=Dropdown(
        options=[
            "Strike Price",
            "Volatility",
            "Time to Maturity",
            "Risk-Free Rate",
        ],
        value="Strike Price",
        description="Varying:",
    ),
    model=Dropdown(
        options=["Black-Scholes", "BOPM"],
        value="Black-Scholes",
        description="Model:",
    ),
)
```

```python
import yfinance as yf
import pandas as pd
import numpy as np
from scipy.stats import norm
from datetime import datetime
import matplotlib.pyplot as plt

# bsm
def bsm_call_price(S, K, T, r, sigma):
    d1 = (np.log(S/K) + (r + 0.5*sigma**2)*T) / (sigma * np.sqrt(T))
    d2 = d1 - sigma * np.sqrt(T)
    call = S * norm.cdf(d1) - K * np.exp(-r*T) * norm.cdf(d2)
    return call

def bsm_greeks(S, K, T, r, sigma):
    d1 = (np.log(S/K) + (r + 0.5*sigma**2)*T) / (sigma * np.sqrt(T))
    d2 = d1 - sigma * np.sqrt(T)

    delta = norm.cdf(d1)
    gamma = norm.pdf(d1) / (S * sigma * np.sqrt(T))
    theta = (-S * norm.pdf(d1) * sigma / (2 * np.sqrt(T))
             - r * K * np.exp(-r*T) * norm.cdf(d2)) / 365
    vega = S * norm.pdf(d1) * np.sqrt(T) / 100
    rho = K * T * np.exp(-r*T) * norm.cdf(d2) / 100

    return delta, gamma, theta, vega, rho

# bopm
def bopm_call_price(S, K, T, r, sigma, N=100):
    dt = T / N
    u = np.exp(sigma * np.sqrt(dt))
    d = 1 / u
    p = (np.exp(r * dt) - d) / (u - d)

    prices = np.array([S * (u ** j) * (d ** (N - j)) for j in range(N + 1)])
    call_values = np.maximum(prices - K, 0)

    for i in range(N - 1, -1, -1):
        call_values = np.exp(-r * dt) * (p * call_values[1:] + (1 - p) * call_values[:-1])

    return call_values[0]

# yfinance
ticker = "NVDA"
stock = yf.Ticker(ticker)
expiry = stock.options[10]
calls = stock.option_chain(expiry).calls.copy()

S = stock.history(period="1d")["Close"][-1]
today = datetime.now().date()
expiry_date = datetime.strptime(expiry, "%Y-%m-%d").date()
T = (expiry_date - today).days / 365
r = 0.05

calls = calls[['strike', 'lastPrice', 'impliedVolatility']].dropna()
calls = calls[(calls['strike'] >= S * 0.8) & (calls['strike'] <= S * 1.2)].reset_index(drop=True)

# computing option prices
bsm_prices = []
bopm_prices = []
deltas, gammas, thetas, vegas, rhos = [], [], [], [], []

for _, row in calls.iterrows():
    K = row['strike']
    sigma = row['impliedVolatility']
    bsm_prices.append(bsm_call_price(S, K, T, r, sigma))
    bopm_prices.append(bopm_call_price(S, K, T, r, sigma, N=100))
    d, g, t, v, rho = bsm_greeks(S, K, T, r, sigma)
    deltas.append(d)
    gammas.append(g)
    thetas.append(t)
    vegas.append(v)
    rhos.append(rho)

# final table
calls['BSM_Price'] = bsm_prices
calls['BOPM_Price'] = bopm_prices
calls['Delta'] = deltas
calls['Gamma'] = gammas
calls['Theta'] = thetas
calls['Vega'] = vegas
calls['Rho'] = rhos
calls.rename(columns={'lastPrice': 'YFinance_Price'}, inplace=True)

# printing table
print(calls[['strike', 'YFinance_Price', 'BSM_Price', 'BOPM_Price', 'Delta', 'Gamma', 'Theta', 'Vega', 'Rho']])

# plotting
plt.figure(figsize=(10,6))
plt.plot(calls['strike'], calls['YFinance_Price'], label="YFinance", marker='o')
plt.plot(calls['strike'], calls['BSM_Price'], label="BSM", marker='x')
plt.plot(calls['strike'], calls['BOPM_Price'], label="BOPM", marker='s')
plt.xlabel("Strike Price")
plt.ylabel("Call Option Price")
plt.title(f"{ticker} Call Option Prices (Expiry: {expiry})")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

#**The Market Reality vs. Model Theories**
There are a number of reasons why the model-derived option prices vary from market prices and all because of mainly the assumptions we made when developing the models.

 Also there are realistic real life reasons for example:
*   News occurs.
*   Liquidity disappears.
* Some options don't trade at all during the day.

This contradictory behavior results in the varying outcomes.

##1. Model Assumptions Do Not Reflect Reality
BSM and BOPM both use assumptions to make the mathematics easier. But actual markets resist those assumptions on a daily basis:

Assumption: What model assumes and what actually occurs
* Constant Volatility
* Smooth & predictable price movements
* Volatility wobbles with news, panic
* No Transaction Costs. You can trade instantly, freely. Real trades have fees & spreads
* Continuous Time Trading. You can trade in every millisecond but gaps, markets close
* No Dividends or Perfectly Known Dividends
* No cash dividends influence prices
* Dividend-paying stocks influence call values
* No Arbitrage
* Prices are always fair. Markets misprice due to fear, greed, or inefficiency
* Log-normal Returns
* Prices have a bell-shaped distribution. Real returns have jumps, crashes

These distinctions result in real distinctions between model values and market prices.

##2. Implied Volatility ≠ Constant Volatility
BSM and BOPM need just one volatility input — but which?

In the real world, volatility depends on strike and expiry.

Example:
A call option with a $100 strike may have 20% IV.

Another with a $80 strike may have 30% IV.

This provides us with what is known as the volatility smile or skew.
The BSM model assumes a flat line.
Markets tell us: "Nope. It's curved."

Therefore, BSM returns to you a price using one input volatility.
But the market price is a snapshot of a whole volatility surface that models cannot capture.

##3. Liquidity, Bid-Ask Spreads & Last Trade Noise
When you pull a price from YFinance, what you're typically looking at is the last trade price — not necessarily a fair price.

* That last trade could have occurred hours before.

* Maybe it was a desperate buyer or seller, and not indicative of fair value.

* Some options never get exercised during the day at all.

BSM provides you with an exact math-of-things estimate of what the price "should" be in a discrete time period where everything is perfect.

So: Market price ≠ Fair value. Perhaps it's simply the outcome of an unbalanced trade.

##4. Dividend Impacts Aren't Covered in Vanilla Models
Assume a company is to distribute a dividend. That dividend lowers the stock price when distributed.

* A call option will lose value, since the stock will drop.
* A put option will gain value, for the same reason.

The vanilla BSM model doesn't account for dividends other than where individually adjusted.
True market prices do take dividends into account.

This results in repeated overcalling of calls by BSM on dividend-paying stocks — unless you apply a dividend-adjusted version of the model.

##5. Early Exercise (American Options) vs. European Models
* The majority of US stock options are American-style:

* May be exercised at any point prior to expiry.

* But BSM is for European-style options:

------------------------------------------------

Early exercise is best when interest is high, or dividend is near-term.
BSM underestimates these opportunities since it excludes early exercise.

Although BOPM in theory can price American options by looking for early exercise at each point, it needs to be coded up and done as a matter of fact — which most straightforward implementations (such as yours or mine) unfortunately do not happen to do as a matter of course.

##6. Human Behavior & Market Psychology
* Markets are made up of human beings (and now human-programmed computers). Sentiment counts.

* In earnings weeks or crises, options become mispriced with fear.

* Individuals overpay for winning lottery tickets (calls) or insurance (puts).

* Real-world risk premiums have a role here. Models such as BSM are unaware of:

* Earnings calls


They simply price off of arbitrary price motion.

##7. Numerical Approximation Errors (particularly in BOPM)
BOPM constructs a price tree in N time steps. But:

* If N is very small, you obtain crude approximations.

* If you don't permit early exercise in your tree, you lose actual American pricing.

* Very high volatilities or very short maturities can lead to instability.

* Although BOPM is more general than BSM, it's more sensitive to the implementation options.
